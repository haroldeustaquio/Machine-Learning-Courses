```python
import numpy as np
from collections import Counter
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# make_blobs()
X, y = make_blobs(n_samples=8000, centers=4, random_state=1)

print(X.shape, y.shape)
print('hola')
```

```python
for label, _ in Counter(y).items():
    temp_label = np.where(y==label)[0]
    
    if label==0:
        plt.scatter(X[temp_label][:,0], X[temp_label][:,1], c='blue')
    elif label==1:
        plt.scatter(X[temp_label][:,0], X[temp_label][:,1], c='red')
    elif label==2:
        plt.scatter(X[temp_label][:,0], X[temp_label][:,1], c='orange')
    else: 
        plt.scatter(X[temp_label][:,0], X[temp_label][:,1],c='black')
```
